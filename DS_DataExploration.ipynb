```python
#1. Clean dirty data

#A. Missing data
#we fill nan values with mean because it's small
df[feature_except_target] = df[feature_except_target].fillna(df[feature_except_target].mean())
print("# of na values: ", df.isna().sum().sum())
```

```python
#B. Outliers


#IQR method to remove outlier
def remove_outliers(df, feature_names):
    for feature in feature_names:
        Q1 = df[feature].quantile(0.25)
        Q3 = df[feature].quantile(0.75)
        IQR = Q3 - Q1
        # generally use 1.5 instead of 2, if so ,# of data become very small, so set 2
        lower_bound = Q1 - 2 * IQR
        upper_bound = Q3 + 2 * IQR

        # fitering
        df = df[(df[feature] >= lower_bound) & (df[feature] <= upper_bound)]
    return df


print("Original DataFrame size:", df.shape)
df = remove_outliers(df, feature_except_target)
print("Cleaned DataFrame size:", df.shape)
```

```python
#2. normalization

from sklearn.preprocessing import MinMaxScaler, StandardScaler, RobustScaler

# # ----------------------------------------------------------------
# # Scaling
# # Input: train Dataset, test Dataset, scaler name to use
# # Output: scaled dataset
# # ----------------------------------------------------------------

def scale(df_train, df_test, scaler_name):
    df_train_scaled = df_train.copy()
    df_test_scaled = df_test.copy()
    #MinMaxScaling
    if scaler_name == 'MinMax':
        scaler = MinMaxScaler()
        scaler.fit(df_train_scaled)
        df_train_scaled = scaler.transform(df_train_scaled)
        df_test_scaled = scaler.transform(df_test_scaled)
    #StandardScaling
    elif scaler_name == 'Standard':
        scaler = StandardScaler()
        scaler.fit(df_train_scaled)
        df_train_scaled = scaler.transform(df_train_scaled)
        df_test_scaled = scaler.transform(df_test_scaled)
    #RobustScaling
    elif scaler_name == 'Robust':
        scaler = RobustScaler()
        scaler.fit(df_train_scaled)
        df_train_scaled = scaler.transform(df_train_scaled)
        df_test_scaled = scaler.transform(df_test_scaled)
    else:
        print("error in scaler")
        return None
    return df_train_scaled, df_test_scaled

#3. Unusable data
#Drop unusable feature - function pair is another target, but we don't use this
df = df.drop(['function_pair'], axis = 1)
```

```python
#3. Encoding for data mining algorithm & Feautre Creation


#Encoding key features
def group_by_note(key):
    if 'C' in key:
        return 'C'
    elif 'D' in key:
        return 'D'
    elif 'E' in key:
        return 'E'
    elif 'F' in key:
        return 'F'
    elif 'G' in key:
        return 'G'
    elif 'A' in key:
        return 'A'
    elif 'B' in key:
        return 'B'
    else:
        return 'Other'



# # -----------------------------------------------------------------------
# # Encoding
# # Input: dataframe, key_features(list), use_encoding and use_note : boolean
# # Output: dataframe
# # -----------------------------------------------------------------------

def encode(df, key_features, use_encoding, use_note):
    if use_encoding:
        df_encoded = df.copy()
        print("Encoding used")
        # Feature creation
        df_encoded['dominant_key'] = df_encoded[key_features].idxmax(axis=1)
        #Drop key features
        df_encoded = df_encoded.drop(key_features, axis = 1)
        # C# -> C, D# -> D, ..
        df_encoded['note'] = df_encoded['dominant_key'].apply(group_by_note)

        # 1: if # exist, other: 0
        df_encoded['half'] = df_encoded['dominant_key'].apply(lambda x: 1. if '#' in x else 0.)

        #minor: 0, major: 1
        df_encoded['scale'] = df_encoded['dominant_key'].apply(lambda x: 1. if 'Major' in x else 0.)

        df_encoded = df_encoded.drop(['dominant_key'], axis = 1)
        if use_note:
            df_encoded = pd.get_dummies(df_encoded, columns=['note'])
            return df_encoded
        else:
            df_encoded = df_encoded.drop(['note'], axis = 1)
            return df_encoded
    else:
        return df
```

```python
#5. Feature reduction
from sklearn.decomposition import PCA

# # -----------------------
# # PCA
# # Input: X_train, X_test(df or np array), use_pca: boolean
# # Output: X_train, X_test
# # -----------------------
def run_pca(X_train, X_test, use_pca):
    if use_pca:
        X_train_pca = X_train.copy()
        X_test_pca = X_test.copy()
        print("PCA used")
        Pca = PCA(.95)
        Pca.fit(X_train_pca)
        X_train_pca = Pca.transform(X_train_pca)
        X_test_pca = Pca.transform(X_test_pca)
        print("Explained variance ratio: ", Pca.explained_variance_ratio_)
        print("Number of components:", Pca.n_components_)
        return X_train_pca, X_test_pca
    else:
        return X_train, X_test
```

```python
#Description of this fuctiom: this function do majority voting
#input predictions must be n_predictions x n_sample
#return: majority vote (shape: (n_prediction,)
def majorityVoting(predictions, n_categories):
    n_predictions = predictions.shape[0]
    n_sample = predictions.shape[1]

    #to store majority votes
    majority_votes = np.zeros((n_sample), dtype=int)

    for i in range(n_sample):
        #count frequency
        counts = np.bincount(predictions[:, i], minlength=n_categories)

        #index from argmax is highest count category
        majority_category = np.argmax(counts)

        #store majority vote
        majority_votes[i] = majority_category

    return majority_votes

```
